```python
import matplotlib.pyplot as plt

#     
files = ["D:/parallel programming/result_2/stats_2_threads.txt",
         "D:/parallel programming/result_2/stats_3_threads.txt",
         "D:/parallel programming/result_2/stats_4_threads.txt",
         "D:/parallel programming/result_2/stats_8_threads.txt",
         "D:/parallel programming/result_2/stats_12_threads.txt",
         "D:/parallel programming/result_2/stats_20_threads.txt"]

#           
sizes_list = []
times_list = []

#    
for file_path in files:
    sizes = []
    times = []
    with open(file_path, 'r') as file:
        for line in file:
            size, time = map(int, line.strip().split('\t'))
            sizes.append(size)
            times.append(time)
    sizes_list.append(sizes)
    times_list.append(times)

#  
plt.figure(figsize=(10, 6))

for i in range(len(files)):
    plt.plot(sizes_list[i], times_list[i], marker='o', linestyle='-', label=f'{[2, 3, 4, 8, 12, 20][i]} ')

plt.title('      ')
plt.xlabel(' ')
plt.ylabel(' ()')
plt.legend()
plt.grid(True)
plt.show()
```
